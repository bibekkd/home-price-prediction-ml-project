## Scikit-learn Design

Primarily three types of objects
1. Estimators - It estimates some parameter based on a dataset. Eg. Imputer
It has a fit method and transform method.
Fit Method - Fits the datasets and calculates internal parameters

2. Transformers - transform method takes input and returns output based on the
learnings from fit(). It also has a conversion function called fit_transform()
which fits and then transform

3. Predictors - LinearRegression model is an example of predictor. fit() and
predict() are two common functions. It alos gives score() function which will evaluate the predictions

## Feature Scaling

Primarily, two types of features scaling methods:
1. Min-max scaling (Normalization)
    (value - min)/(max - min)
   Sklearn provides a class called MinMaxScaler for this 

2. Standardization
   (value - mean)/std
   Sklearn provides a class called StandardScaler for this

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
#it takes list of pipelines you can add as many as you want

my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    # ... add as many as you want in your pipeline
    ('std_scaler', StandardScaler())
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr.shape
```

## Selecting a desired model 

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
some_features = housing[49:54]
```

```python
some_labels = housing_labels.iloc[49:54]
```

```python
prepared_data = my_pipeline.transform(some_features)
prepared_data.shape
```

```python
model.predict(prepared_data)
```

```python
np.array(some_labels)
```

## Evaluating the model

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels, housing_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

## Using better evaluation - Cross Validation 

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, housing_labels, 
                         scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard Deviation:", scores.std())
```

```python
print_scores(rmse_scores)
```

## Testing the model in test data

```python
X_test = strat_test_set.drop("MEDV", axis=1)
y_test = strat_test_set["MEDV"].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
```
